<strong>Название Кейса</strong><br>
<div style="padding-left: 20px; margin-top: 12px;">
Многоклассовое прогнозирование риска ожирения
</div>
</div>

<div class="border-left-dark ">
<strong>Краткое описание кейса</strong><br>
<div style="padding-left: 20px; margin-top: 12px;">
Разработка модели машинного обучения, которая определяет уровень риска ожирения у человека на основе его характеристик (возраст, пол, уровень активности, питание)
</div>
</div>


<div class="border-left-dark ">
<strong>Ожидаемый результат</strong><br>
<div style="padding-left: 20px; margin-top: 12px;">
Создание модели многоклассовой классификации для предсказания категории риска ожирения (низкий, средний, высокий)
</div>
</div>


<div class="border-left-dark ">
<strong>Основной функционал</strong><br>
<ul>
  <li>Очистка и предобработка данных (обнаружение пропусков, нормализация, кодирование категориальных переменных)</li>
  <li>Анализ корреляций между факторами</li>
  <li>Обучение моделей: SVM, Random Forest, градиентный бустинг</li>
  <li>Оценка качества моделей (метрики: accuracy, confusion matrix, macro F1-score)</li>
</ul>
</div><br>
