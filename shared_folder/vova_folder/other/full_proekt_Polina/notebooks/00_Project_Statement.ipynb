<div class="border-left-dark ">
<strong>Проблема</strong><br>
<div style="padding-left: 20px; margin-top: 12px;">
Рост заболеваемости ожирением во всем мире представляет серьезную угрозу для общественного здоровья. Отсутствие эффективных инструментов для раннего выявления групп риска приводит к поздней диагностике, развитию сопутствующих заболеваний (диабет,сердечно-сосудистые патологии) и увеличению нагрузки на систему здравоохранения. Существующие методы оценки риска часто основаны на ограниченном наборе параметров и не учитывают комплексное влияние образа жизни, пищевых привычек и генетических факторов.  
</div>
</div>

<div class="border-left-dark ">
<strong>Цель проекта</strong><br>
<div style="padding-left: 20px; margin-top: 12px;">
Разработать модель машинного обучения для многоклассовой классификации уровня риска ожирения на основе комплексного анализа антропометрических, поведенческих и диетологических данных. Создать инструмент для раннего выявления пациентов с различными степенями риска ожирения для своевременного профилактического вмешательства.
</div>
</div>

<div class="border-left-dark ">
<strong>Ожидаемые результаты</strong><br>
<ol>
  <li>Обученная и валидированная модель классификации с точностью не менее 85%</li>
  <li>Аналитический отчет о значимости различных факторов риска</li>
  <li>Документация по использованию модели и интерпретации результатов</li>
  <li>Рекомендации по профилактическим мероприятиям для каждой группы риск</li>
</ol>
</div>


<div class="border-left-dark ">
<strong>Пользователи/заказчики</strong><br>
<ol>
  <li>Врачи-эндокринологи и диетологи - для скрининга пациентов и разработки персонализированных программ</li>
  <li>Медицинские клиники и диагностические центры - для массового обследования населения</li>
  <li>Корпоративные wellness-программы - для оценки здоровья сотрудников</li>
  <li>Научно-исследовательские институты - для эпидемиологических исследований</li>
  <li>Государственные органы здравоохранения - для планирования профилактических мероприятий</li>
</ol>
</div>


<div class="border-left-dark ">
<strong>Критерии успешности</strong><br>
<ol>
  <li>Модель демонстрирует accuracy > 85% и F1-score > 0.8 на тестовой выборке</li>
  <li>Время предсказания для одного пациента не превышает 5 секунд</li>
  <li>Интерфейс интуитивно понятен для медицинских работников без технической подготовки</li>
  <li>Проект защищен с положительной оценкой экспертов</li>
  <li>Получены положительные отзывы от тестовых пользователей (врачей)</li>
</ol>
